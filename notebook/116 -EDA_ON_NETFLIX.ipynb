## Data Cleaning

```python
df['rating'].value_counts()
```

---                          ---The ratings data has some some incorect entries that we need to set

```python
df['rating'] = df['rating'].replace({'74 min':'TV-MA', '84 min': 'TV-MA', '66 min': 'TV-MA'})
df['rating'] = df['rating'].replace({'TV-Y7-FV': 'TV-V7'})
df['rating'].unique()
```

```python
# Renaming values for Rating for better understanding
df["rating"] = df["rating"].replace({
    "PG-13": "Teens (13+)",
    "TV-MA": "Adults",
    "PG": "Kids (Parental Guidance)",
    "TV-14": "Teens (14+)",
    "TV-PG": "Kids (Parental Guidance)",
    "TV-Y": "Kids",
    "TV-Y7": "Kids (7+)",
    "R": "Adults",
    "TV-G": "Kids",
    "G": "Kids",
    "NC-17": "Adults",
    "NR": "Not Rated",
    "TV-Y7-FV": "Kids (7+)",
    "UR": "Unrated"
})
```

```python
df['rating'] = df['rating'].replace({
    'TV-V7': 'Kids (7+)',
    'Unrated': 'Not Rated',
    'A': 'Adults'
})
```

```python
df.rating.unique()
```

```python
df.country =df.country.fillna(df.country.mode()[0])
```

```python
df.country.isna().sum()
```

```python
#df.cast.replace(np.nan,'None',inplace =True)
#df.director.replace(np.nan,'None',inplace =True)
df['cast'] = df['cast'].replace(np.nan, 'None')
df['director'] = df['director'].replace(np.nan, 'None')
df.dropna(inplace= True)
df.drop_duplicates(inplace= True)
df.head()
```

```python
#Let's drop the column that we don't need Ltke show_id and etc
df_clean = df.drop(["show_id","cast",'description'], axis = 1)
df_clean.head()
```

### Question 2
          Which countries produce the most Netflix content?

```python
##  Splitting the country into different rows
country_df = df_clean[['title','type','country']]
country_df = (country_df.drop('country',axis=1)
             .join(country_df.country.str.split(", ",expand=True).stack().reset_index(drop=True, level=1).rename('country')))
country_df.head()
```

```python
country_df.country.nunique()                        ## Its showing that 127 unique values are there in country 
```

```python
#as we can see we hove records from 127 different countries, we'll onty work wtth top 10 highest contributing countries
top10_countries = country_df['country'].value_counts().sort_values(ascending=False)[:10]
top10_countries
```

```python
x_countries = top10_countries.index
y_countries = top10_countries.values
plt.figure(figsize=(15,8))
plt.title("Top 10 Countries Producing Shows")
plt.xlabel("Countries")
plt.ylabel("Number of shows")
plt.xticks(rotation = 45)
sns.barplot(x =x_countries, y = y_countries,hue= x_countries,palette = 'rocket',legend =False)
plt.show()
```

#### Takeaway: As we can see that the United States, India, and United Kingtom are the top contributing countries.

```python
temp_df =country_df.loc[country_df['country'].isin(top10_countries.index)]
temp_df.head()
```

```python
#iet's see side by side compariston
plt.figure(figsize=(17,6))
plot = sns.countplot(x ='type',data=temp_df,hue="country",palette='husl')
plot.set_ylabel("Count of Content",fontsize=10, weight='bold')
plot.set_xlabel("Countries",fontsize=10,weight='bold')
plt.legend(title="country", loc='upper right')
theme = ["#DEFSE5", "#459E97", "#489ABB", "#044361", "#E68193"]
plot.figure.suptitle(" Comparision of TV Show/ Movies of top 10 content producing countries",y=1, size = 20, color = theme[3],weight ='bold')
```

```python
movie_countries = country_df[country_df['type']=='Movie']
tv_show_countries = country_df[country_df['type']=='TV Show']
```

```python
top10_movies_countries = movie_countries.country.value_counts().sort_values(ascending= False)[:10]
top10_tv_shows_countries = tv_show_countries.country.value_counts().sort_values(ascending=False)[:10]
```

```python
plt.subplots(figsize=(12,6))
sns.barplot(x=top10_movies_countries.index, y=top10_movies_countries.values,palette="cubehelix")
plt.title("Top 10 countries producing movies in Netflix")
plt.xlabel("Countries", fontsize=10, weight='bold')
plt.ylabel("Count of Movies", fontsize=10, weight="bold")
plt.xticks(rotation=45)
plt.show()


plt.subplots(figsize=(12,6))
sns.barplot(x=top10_tv_shows_countries.index, y=top10_tv_shows_countries.values, palette="flare")
plt.title("Top 10 countries producing TV Shows in Netflix")
plt.xlabel("Countries", fontsize=10, weight='bold')
plt.ylabel("Count of Movies", fontsize=10, weight="bold")
plt.xticks(rotation=45)
plt.show()
```

#### Takeaway: As we can see that the united states is the highest tv show and movie both producing countries.

### Movie Based Genre

```python
df_temp = df_clean.query("type == 'Movie'")
# splitting the Genre in different rows to visualize it
df_temp = df_temp[['title','type', 'country',"listed_in"]]
df_temp = (df_temp.drop('listed_in', axis=1)
           .join(df_temp.listed_in.str.split(",",expand=True).stack().reset_index(drop=True, level=1).rename("Genre")))
df_temp.head()
```

```python
df_temp.Genre.unique()
```

```python
df_temp.Genre.unique().shape
```

```python
df_temp.Genre.unique().size
```

```python
Genre_df = pd.DataFrame(df_temp.Genre.value_counts().sort_values(ascending=False) ).reset_index()
Genre_df.columns = ['Genre','Counts']
Genre_df.head()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

Genres = Genre_df['Genre'].dropna()
plt.figure(figsize=(10,10))
wordcloud = WordCloud(background_color='white',width=500,height=250).generate(' '.join(Genres))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
plt.subplots(figsize=(20,6))
sns.barplot(x=Genre_df["Genre"],y=Genre_df["Counts"],palette="Blues_d")
plt.title('Movie Content in Netflix')
plt.xlabel("Genres")
plt.ylabel("Count of comtent")
plt.xticks(rotation= 90)
plt.show()
```

#### Result: International Movies have been the most popular genre for Movie category followed by Dramas and Comedies
